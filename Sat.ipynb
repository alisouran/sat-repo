# SAT ALGORITHM
## JAHAT VARED KARDAN SOAL AZ STANDARD ZIR ESTEFADE KONID:

**1. BARAY NAGHIZ AZ ALAMAT: '* YA - YA ~' ESTEFADE KONID**
**2. BARAY AND AZ: '& YA AND YA and YA And' ESTEFADE KONID**
**3. BARAY OR AZ: 'OR YA Or YA or' ESTEFADE KONID**
**4. BE FASELE GOZARI HA DEGHAT KONID**

## Algorithm be sorat zir amal mikonad:
**1. TAMAM LITERAL HA DAR LIST SAVE MISHAVAD**
**2. GOZAREHA AZ LITERAL HA ESTEKHRAJ MISHAVAD **
**3. BE GOZAREHA MEGHDAR SEFR DADE MISHAVAD VA DAR DICTUNARY SAVE MISHAVAD **
**4. GOZARE HAYE TANHA MEGHDAR 1 MIGIRAD**
**5. GOZAREHAYE BE SORAT A-->B CHECK MISHAVAD, AGAR TRUE BASHAD GOZARE CHAP MISHAVAD VA BE MARHALE BAAD MIRAVAD, AGAR NA FALSE CHAP MISHAVAD VA UNSAT**
**6. GOZAREHAYE BE SORAT A-->~B CHECK MISHAVAD, AGAR TRUE BASHAD MIRAVAD MARHALE BAAD, AGAR NA FALSE VA UNSAT**
**7. EBTEDA GOZAREHAYE BE SORAT A AND B --> C CHECK MISHAVAD SEPAS GOZAREHAYE BE SORAT A AND B AND C --> 0 CHECK MISHAVAD, AGAR SAT BASAHD PRINT MIKONAD SAT, AGAR NA UNSAT**

## **DEVELOPED BY ALIREZA SOURAN AND SARA NOURI**

```python
# ! FORMAT GOZARE BE HALAT STANDARD

def standardFormat(gozare):
    gozare = gozare.replace("\"", "")
    gozare = gozare.replace("' ", "")
    gozare = gozare.replace(" '", "")
    gozare = gozare.replace("'", "")
    gozare = gozare.replace("or" , "OR")
    gozare = gozare.replace("and", "AND")
    gozare = gozare.replace("And", "AND")
    gozare = gozare.replace("&" , "AND")
    gozare = gozare.replace("Or", "OR")
    gozare = gozare.replace("-" , "~")
    gozare = gozare.replace("*", "~")
    return gozare


# ! TAMAM GOZARE HAYE SADE RA RETURN MIKONAD
def getGozareSade (gozare) :
    gozare = gozare.translate({ord(i): None for i in '~()'})
    gozare = gozare.replace(" ", "")
    gozare = gozare.replace("OR", "")
    gozare = gozare.replace("AND", "")
    arr = []
    for i in gozare:
        arr.append(i)
    arr = list(dict.fromkeys(arr))    
    return arr


# ! TAMAM GOZARE HAYE SADE RA 0 MIKONAD
def setZeroGozareSade(arr):
    gozare_sade={}
    for item in arr:
        gozare_sade[item] = 0
    return gozare_sade


# ! TEDAD NAGHIZ HAYE LITERAL
def getNaghiz(literal):
    for item in literal:
        if(item != "~"):
            literal = literal.replace(item, "")
    return literal


# ! MEGHDAR GOZAREHAYE TANHA
def getGozareTanha(literals, gozare_sefr_shode):
    for item in literals:
        if(len(getNaghiz(item)) == 0):
            gozare_sefr_shode[item.translate({ord(i): None for i in '~()'})] = 1
    return gozare_sefr_shode


# ! CHECK KARDAN SHART 3A
def checkShartSevom(literals, gozarehaye_tanha):
    for item in literals:
        if(len(getNaghiz(item)) >= 1 and len(getGozareSade(item)) > 1):
            if(len(getNaghiz(item)) != len(getGozareSade(item))):
                if(len(getGozareSade(item))==2):
                    item = item.translate({ord(i): None for i in '()'})
                    item = item.split(" OR ")
                    item.sort(key=len, reverse=True)
                    if(gozarehaye_tanha[item[0].replace("~", "")] == 1):
                        gozarehaye_tanha[item[1]] = 1
    for item2 in literals:
        if(len(getNaghiz(item2)) >= 1 and len(getGozareSade(item2)) > 1):
            if(len(getNaghiz(item2)) != len(getGozareSade(item2))):
                if(len(getGozareSade(item2))==2):
                    item2 = item2.translate({ord(i): None for i in '()'})
                    item2 = item2.split(" OR ")
                    item2.sort(key=len, reverse=True)
                    if(gozarehaye_tanha[item2[0].replace("~", "")] != gozarehaye_tanha[item2[1]]):
                        return False
    return gozarehaye_tanha
                        

# ! CHECK SHART BEIN 3B
def checkShartSevomB(literals, gozarehaye_tanha):
    for item in literals:
        if(len(getNaghiz(item)) >= 1 and len(getGozareSade(item)) > 1):
            if(len(getNaghiz(item)) != len(getGozareSade(item))):
                if(len(getGozareSade(item))>2):
                    item = item.translate({ord(i): None for i in '()'})
                    item = item.split(" OR ")
                    item.sort(key=len, reverse=True)
                    for ch in item:
                        if(gozarehaye_tanha[item[-1]] != gozarehaye_tanha[ch.replace("~", "")]):
                            return False
    return True
                
                
# ! CHECK KARDAN SHART 4
def checkShartChaharom (literals, gozarehaye_tanha):
    for item in literals:
        if(len(getNaghiz(item)) >= 1 and len(getGozareSade(item)) > 1):
            if(len(getNaghiz(item)) == len(getGozareSade(item))):
                gozareha = getGozareSade(item)
                dictGozareha = {}
                for item in gozareha:
                    dictGozareha[item] = gozarehaye_tanha[item]
                for item2 in dictGozareha:
                    if(dictGozareha[item2] != 0):
                        return False
                    
    return True




#####!!!<<<--- MAIN --->>>!!!#####

print("\n")
gozare = "(~A OR ~B OR ~C) AND (~E) AND (~C OR A) AND (C) AND (B) AND (~G OR D) AND (G)"
gozare = standardFormat(gozare)



print("\nGOZARE NEMONE VARED SHODE: ")
print(gozare)
print("\n")

# ! GET ALL LITERALS IN A LIST
print("TAMAM LITERAL HA DAR LIST SAVE MISHAVAD: ")
literals = gozare.split(" AND ")
print(literals)
print("\n")


# ! TABE GOZARE HAYE SADE
print("GOZAREHA AZ LITERAL HA ESTEKHRAJ MISHAVAD: ")
gozare_sade = getGozareSade(gozare)
print(gozare_sade)
print("\n")


# ! SEFR KARDAN GOZAREHAYE

print("BE GOZAREHA MEGHDAR SEFR DADE MISHAVAD VA DAR DICTUNARY SAVE MISHAVAD: ")
gozare_sefr_shode = setZeroGozareSade(gozare_sade)
print(gozare_sefr_shode)
print("\n")


# ! DARYAFT TAMAM GOZAREHAYE TANHA BA MEGHDAR
print("GOZARE HAYE TANHA MEGHDAR 1 MIGIRAD: ")
gozarehaye_tanha = getGozareTanha(literals, gozare_sefr_shode)
print(gozarehaye_tanha)
print("\n")


# ! CHECK SHART 3 (GOZAREHAYE BE SORAT A-->B)
print("GOZAREHAYE BE SORAT A-->B CHECK MISHAVAD \nAGAR TRUE BASHAD GOZARE CHAP MISHAVAD VA BE MARHALE BAAD MIRAVAD, AGAR NA FALSE CHAP MISHAVAD: ")
gozarehaye_tanha = checkShartSevom(literals, gozarehaye_tanha)
print(gozarehaye_tanha)
print("\n")




# ! CHECK SAT OR UNSAT (SHART 3B AND SHART 4)
if(gozarehaye_tanha != False):
    print("EBTEDA GOZAREHAYE BE SORAT A AND B --> C CHECK MISHAVAD SEPAS GOZAREHAYE BE SORAT A AND B AND C --> 0 CHECK MISHAVAD \n \nAGAR SAT BASAHD PRINT MIKONAD SAT, AGAR NA UNSAT: \n")
    if(checkShartSevomB(literals, gozarehaye_tanha)):
            if(checkShartChaharom(literals, gozarehaye_tanha)):
                print("SAT")
            else:
                print("UNSAT")
    else:
        print("UNSAT")
else:
    print("UNSAT")
```
